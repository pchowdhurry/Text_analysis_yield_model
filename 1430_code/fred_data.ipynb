```python
# converting to numpy matrix 
# General rule: pick series with largest length, set 
# it as the shape and for series that are shorter, fill in gaps with 0

def convert_np_mat(list_series:list[pd.Series])-> np.ndarray:
    # getting the max length of the series list
    max_size = 0 
    max_index= 0
    count = 0 
    for series in list_series:
        curr_shape = series.shape[0]
        if curr_shape > max_size:
            max_size = curr_shape
            max_index = count
        count+=1 

    # Creating np array with rows as max_size and then adding in data
    # the missing pieces of data are zeros  
    output_mat = np.zeros(shape=(max_size, len(list_series)))
    for i in range(0, len(list_series)):
        curr_data = list_series[i]
        output_mat[0:len(curr_data),i] = curr_data

    return (output_mat, max_index)

```
